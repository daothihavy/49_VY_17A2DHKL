```python
import pandas as pd

# 1. Đọc dữ liệu từ tập tin drinks.csv với index_col là cột đầu tiên
drinks = pd.read_csv("drinks.csv", index_col=0)

# a) Cho biết kiểu dữ liệu (type) và kích thước (shape)
print("Type of 'drinks':", type(drinks))
print("Shape of 'drinks':", drinks.shape)

# b) Hiển thị tên các cột
print("\nColumns of 'drinks':", drinks.columns.tolist())

# c) Xem 5 dòng dữ liệu đầu tiên (head) và cuối cùng (tail)
print("\nFirst 5 rows:")
print(drinks.head())
print("\nLast 5 rows:")
print(drinks.tail())

# 2. Số lượng bia tiêu thụ trung bình ở mỗi châu lục
average_beer_per_continent = drinks.groupby("continent")["beer_servings"].mean()
print("\nAverage beer consumption per continent:")
print(average_beer_per_continent)

# 3. Thống kê tổng quát (describe) số lượng rượu vang tiêu thụ ở mỗi châu lục
wine_stats_per_continent = drinks.groupby("continent")["wine_servings"].describe()
print("\nWine consumption statistics per continent:")
print(wine_stats_per_continent)

# 4. Số lượng bia và rượu tiêu thụ trung bình ở mỗi châu lục
average_beer_wine_per_continent = drinks.groupby("continent")[["beer_servings", "wine_servings"]].mean()
print("\nAverage beer and wine consumption per continent:")
print(average_beer_wine_per_continent)

# 5. Giá trị trung vị (median) cho các loại bia và rượu tiêu thụ ở mỗi châu lục
median_beer_wine_per_continent = drinks.groupby("continent")[["beer_servings", "wine_servings"]].median()
print("\nMedian beer and wine consumption per continent:")
print(median_beer_wine_per_continent)

# 6. Rượu mạnh (spirit_servings) tiêu thụ trung bình, lớn nhất và nhỏ nhất ở mỗi châu lục
spirit_stats_per_continent = drinks.groupby("continent")["spirit_servings"].agg(["mean", "max", "min"])
print("\nSpirit consumption (mean, max, min) per continent:")
print(spirit_stats_per_continent)

# 7. Sắp xếp dữ liệu tăng dần theo số lượng bia tiêu thụ
sorted_by_beer = drinks.sort_values("beer_servings")
print("\nTop 5 countries with the highest beer consumption:")
print(sorted_by_beer.tail(5))

print("\nTop 5 countries with the lowest beer consumption:")
print(sorted_by_beer.head(5))
```

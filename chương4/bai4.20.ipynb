```python
import pandas as pd

# Tải dữ liệu từ file euro12.csv
euro12 = pd.read_csv('euro12.csv')

# 1. In kiểu dữ liệu (type), kích thước (shape) và danh sách các cột
print("Type:", type(euro12))
print("Shape:", euro12.shape)
print("Columns:", euro12.columns.tolist())

# 2. In giá trị cột "Goals"
print("\nGoals column values:")
print(euro12['Goals'])

# 3. Có bao nhiêu đội tham gia Euro 2012?
num_teams = euro12['Team'].nunique()
print("\nNumber of teams:", num_teams)

# 4. In thông tin tổng quan của DataFrame
print("\nEuro2012 Information:")
print(euro12.info())

# 5. Tạo DataFrame discipline chỉ chứa 3 cột 'Team', 'Yellow Cards', 'Red Cards'
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# 6. Sắp xếp discipline giảm dần theo 2 cột 'Red Cards', 'Yellow Cards'
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\nDiscipline sorted:")
print(discipline_sorted)

# 7a. Tính trung bình 'Yellow Cards'
yellow_cards_mean = euro12['Yellow Cards'].mean()
print("\nAverage Yellow Cards:", yellow_cards_mean)

# 7b. Lọc các đội đã ghi hơn 6 bàn thắng
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]
print("\nTeams with more than 6 goals:")
print(teams_more_than_6_goals)

# 8. In các đội mà tên bắt đầu bằng chữ 'G'
teams_start_with_g = euro12[euro12['Team'].str.startswith('G')]
print("\nTeams starting with 'G':")
print(teams_start_with_g)

# 9. In 7 cột đầu của euro12
print("\nFirst 7 columns of euro12:")
print(euro12.iloc[:, :7])

# 10. In tất cả các cột, trừ 3 cột cuối
print("\nAll columns except the last 3:")
print(euro12.iloc[:, :-3])

# 11. In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print("\nSelected columns:")
print(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])

# 12. In các cột 'Team', 'Shooting Accuracy' chỉ cho 'England', 'Italy', 'Russia'
selected_teams = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]
print("\nSelected teams' Shooting Accuracy:")
print(selected_teams[['Team', 'Shooting Accuracy']])
```

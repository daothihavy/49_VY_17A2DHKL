```python
import pandas as pd
import numpy as np

# 1. Đọc và kiểm tra dữ liệu từ các file
# a) Đọc stocks1.csv
stocks1 = pd.read_csv("stocks1.csv")
print("First 5 rows of stocks1:")
print(stocks1.head())
print("\nLast 5 rows of stocks1:")
print(stocks1.tail())
print("\nDtypes of stocks1:")
print(stocks1.dtypes)
print("\nInfo of stocks1:")
print(stocks1.info())

# b) Đọc stocks2.csv
stocks2 = pd.read_csv("stocks2.csv")
print("\nFirst 5 rows of stocks2:")
print(stocks2.head())
print("\nLast 5 rows of stocks2:")
print(stocks2.tail())
print("\nDtypes of stocks2:")
print(stocks2.dtypes)
print("\nInfo of stocks2:")
print(stocks2.info())

# c) Đọc companies.csv
companies = pd.read_csv("companies.csv")
print("\nData of companies:")
print(companies.head())
print("\nDtypes of companies:")
print(companies.dtypes)
print("\nInfo of companies:")
print(companies.info())

# 2. Kiểm tra và xử lý giá trị Null trong stocks1
print("\nAre there Null values in stocks1?")
print(stocks1.isnull().sum())

# Xử lý các giá trị Null
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))
print("\nNull values after filling in stocks1:")
print(stocks1.isnull().sum())

# 3. Gộp stocks1 và stocks2 theo dòng
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print("\nLast 15 rows of combined stocks:")
print(stocks.tail(15))

# 4. Gộp stocks và companies
stocks_companies = stocks.merge(companies, left_on="symbol", right_on="name", how="inner")
print("\nFirst 5 rows of stocks_companies:")
print(stocks_companies.head())

# 5. Tính giá (open, high, low, close) trung bình và volume trung bình của mỗi công ty
average_prices = stocks.groupby("symbol")[["open", "high", "low", "close", "volume"]].mean()
print("\nAverage prices and volume per company:")
print(average_prices)

# 6. Tính giá đóng cửa trung bình, lớn nhất và nhỏ nhất của mỗi công ty
close_stats = stocks.groupby("symbol")["close"].agg(["mean", "max", "min"])
print("\nClose price statistics (mean, max, min) per company:")
print(close_stats)

# 7. Tạo cột parsed_time và chuyển sang định dạng DateTime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print("\nDtype of parsed_time column:", stocks_companies['parsed_time'].dtype)
print("\nFirst 5 rows with parsed_time:")
print(stocks_companies.head())

# 8. Thêm cột result
stocks_companies['result'] = np.where(stocks_companies['close'] > stocks_companies['open'], 'up', 'down')
print("\nFirst 5 rows with result column:")
print(stocks_companies.head())
```
